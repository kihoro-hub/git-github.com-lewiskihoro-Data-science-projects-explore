```python
#finding average of age of students

passengers = [
    {"name": "Alice","age": 29,"survived": 1},
    {"name": "Bob","age": 21,"survived": 0},
    {"name": "Charlie","age": 30,"survived": 1}
]
```

```python
ages =[p["age"] for p in passengers if p["age"]is not None]
average_age = sum(ages)/len(ages)
print(f"Average age of passengers: {average_age:.2f}")
```

```python
import pandas as pd
df=pd.DataFrame(passengers)
df
```

```python
average=df['age'].mean()
average
```
