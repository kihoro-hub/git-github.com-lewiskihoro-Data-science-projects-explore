```python
class LibraryItem:

    _total_items = 0  

    def __init__(self, title, author, publication_year,is_available=True):
        self.title = title
        self.author = author
        self.publication_year = publication_year
        self.is_available = is_available
        self.checkout_history = []
        LibraryItem._total_items += 1

    def display_info(self):
        availability = 'Available' if self.is_available else 'Checked Out'
        print(f'Title: {self.title}')
        print(f'Author: {self.author}')
        print(f'Publication Year: {self.publication_year}')
        print(f'Status: {availability}')

    def checkout(self, librarian_name):
        if not self.is_available:
            print(f"Sorry, '{self.title}' is currently checked out.")
            return False
        self.is_available = False
        self.checkout_history.append({'action': 'checkout', 'librarian': librarian_name})
        print(f"You have checked out '{self.title}'.")
        return True

    def return_item(self, librarian_name):
        if self.is_available:
            print(f"'{self.title}' is not currently checked out.")
            return False
        self.is_available = True
        self.checkout_history.append({'action': 'return', 'librarian': librarian_name})
        print(f"You have returned '{self.title}'.")
        return True

    def show_checkout_history(self):
        if not self.checkout_history:
            print(f"No checkout history for '{self.title}'.")
            return
        print(f"Checkout History for '{self.title}':")
        for entry in self.checkout_history:
            print(f"  - {entry['action'].capitalize()} by {entry['librarian']}")
```

```python
class Book(LibraryItem):

    def __init__(self, title, author, publication_year, isbn, publisher):
        super().__init__(title, author, publication_year)
        self.isbn = isbn
        self.publisher = publisher

    def display_info(self):
        print(f"\n{'='*40}")
        print(f'  BOOK')
        print(f"{'='*40}")
        super().display_info()
        print(f'ISBN: {self.isbn}')
        print(f'Publisher: {self.publisher}')
        print(f"{'='*40}\n")
```

```python
class Magazine(LibraryItem):

    def __init__(self, title, author, publication_year, issue_number, publication_month):
        super().__init__(title, author, publication_year)
        self.issue_number = issue_number
        self.publication_month = publication_month

    def display_info(self):
        print(f"\n{'='*40}")
        print(f'  MAGAZINE')
        print(f"{'='*40}")
        super().display_info()
        print(f'Issue Number: {self.issue_number}')
        print(f'Publication Month: {self.publication_month}')
        print(f"{'='*40}\n")
```

```python
class DVD(LibraryItem):

    def __init__(self, title, publication_year, duration, director, genres, author='N/A'):
        super().__init__(title, author, publication_year)
        self.duration = duration
        self.director = director
        self.genres = set(genres)

    def add_genre(self, genre):
        self.genres.add(genre)
        print(f"Genre '{genre}' added to '{self.title}'.")

    def display_info(self):
        print(f"\n{'='*40}")
        print(f'  DVD')
        print(f"{'='*40}")
        super().display_info()
        print(f'Director: {self.director}')
        print(f'Duration: {self.duration} minutes')
        print(f"Genres: {', '.join(sorted(self.genres))}")
        print(f"{'='*40}\n")
```

```python

print('--- Creating items ---\n')

book = Book(
    title='Children Of Blood and Bone',
    author='Tomi Adeyemi',
    publication_year=2003,
    isbn='109-490678900',
    publisher='Henry Holt and Company'
)

magazine = Magazine(
    title='Vogue',
    author='Multiple Authors',
    publication_year=2026,
    issue_number=6,
    publication_month=4
)

dvd = DVD(
    title='The Wolf of Wall Street',
    publication_year=2013,
    duration=180,
    director='Martin Scorsese',
    genres={'Thriller', 'Comedy'}
)
#  TEST display_info

print('--- Display Info ---')
book.display_info()
magazine.display_info()
dvd.display_info()

#  TEST checkout and return_item

print('--- Checkout & Return ---\n')

# Check out the book
book.checkout(librarian_name='Alicia Mwangi')
print(f'is_available after checkout: {book.is_available}')   # Expected: False

# Try to check out again 
book.checkout(librarian_name='Jane Doe')

# Return the book
book.return_item(librarian_name='Alicia Mwangi')
print(f'is_available after return:   {book.is_available}')   # Expected: True

# Try to return again 
book.return_item(librarian_name='Alicia Mwangi')

# Show checkout history
book.show_checkout_history()


#  TEST add_genre

print('\n--- DVD add_genre ---\n')

print(f'Genres before: {dvd.genres}')   # Expected: {'Thriller', 'Comedy'}
dvd.add_genre('Drama')                  # new genre — should be added
dvd.add_genre('Comedy')                 # duplicate — silently ignored by set
print(f'Genres after:  {dvd.genres}')   # Expected: {'Thriller', 'Comedy', 'Drama'}

# ============================================================
#  TOTAL ITEMS CREATED
# ============================================================
print(f'\nTotal LibraryItem instances ever created: {LibraryItem._total_items}')
```
