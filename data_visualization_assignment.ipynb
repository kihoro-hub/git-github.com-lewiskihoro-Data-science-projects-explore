```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Barchart

```python
height =[40,30,10,50,25,5]
x = [0,1,2,3,4,5]
labels = ["Crime","Drama","Comedy","Action","Sci-fi","Documentary"]

fig, ax = plt.subplots(figsize=(8, 6))

ax.bar(x, height, color='blue', tick_label=labels)

ax.set_title("Jim's Video library")
ax.set_xlabel("Genres")
ax.set_ylabel("Number of movies")


```

Scatterplot

```python

weight = [2750, 3125, 2100, 4082, 2690, 3640, 4380, 2241, 2895, 3659]
mpg = [29, 23, 33, 28, 20, 21, 14, 25, 31, 17]

fig ,ax=plt.subplots(figsize=(8, 6))

ax.scatter(weight, mpg, color='blue',label ="weight vs mileage")

ax.set_xlabel("car weight")
ax.set_ylabel("miles per gallon")
ax.set_title("Car weight vs miles per gallon")
ax.legend()
plt.show()
```

Histogram

```python
x = [43.1, 35.6, 37.5, 36.5, 45.3, 43.4, 
     40.3, 50.2, 47.3, 31.2, 42.2, 45.5, 
     30.3, 31.4, 35.6, 45.2, 54.1, 45.6, 
     36.5, 43.1]
fig,ax = plt.subplots(figsize=(8, 6))
ax.hist(x,bins=5, color='tab:blue')
ax.set_title("Customer waiting times")
ax.set_xlabel("Waiting time (minutes)")
ax.set_ylabel("Number of customers")
plt.show()
```
