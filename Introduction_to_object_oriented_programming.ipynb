Inheritance

```python
-It's a principle that allows one class(child) to inherit attributes and methods from another class(parent/super)<br>
-This means the subclass inherits properties and behaviors of the superclass but can still define its own properties.<br>
Example:<br>
>we can have two bank clients,one is a regular client and the other is a premium client.<br>
>There is a section of the bank that is only for premium clients.<br>
If we were to programme that we have two classes,one for regular clients and one for premium clients.The latter will inherit from the super.<br>
All clients have : name,account number,account balance e.t.c<br>
```

```python
class BankClient:
    def __init__(self,name,account_number):
        self.name= name
        self.account= account_number

class PremiumClient:
    def __init__(self,name, email,account_number,stock_portfolio):
        super().__init__(name,account_number,email)
        self.stock = stock_portfolio 
        self.email = email
```
